```python
import matplotlib.pyplot as plt

# 数据
k_values = [2, 4, 6, 8, 10]
DoS = [1.00, 1.00, 1.00, 1.00, 1.00]
Probe = [0.99, 0.99, 1.00, 1.00, 1.00]
U2R = [0.54, 0.77, 0.79, 0.80, 0.75]
R2L = [0.88, 0.92, 0.95, 0.96, 0.98]
Normal = [0.99, 1.00, 1.00, 1.00, 1.00]

# 颜色选择
colors = {
    "DoS": "dodgerblue",
    "Probe": "crimson",
    "U2R": "olive",
    "R2L": "darkorange",
    "Normal": "deepskyblue"
}

# 画图
plt.figure(figsize=(6, 5), facecolor='white')  # 设置白色背景
plt.plot(k_values, [x * 100 for x in DoS], 'o-', color=colors["DoS"], label="DoS")
plt.plot(k_values, [x * 100 for x in Probe], 'd-', color=colors["Probe"], label="Probe")
plt.plot(k_values, [x * 100 for x in U2R], 's-', color=colors["U2R"], label="U2R")
plt.plot(k_values, [x * 100 for x in R2L], 'p-', color=colors["R2L"], label="R2L")
plt.plot(k_values, [x * 100 for x in Normal], 'h-', color=colors["Normal"], label="Normal")

# 坐标轴与标签
plt.xlabel("K-value", fontsize=12)
plt.ylabel("DR(%)", fontsize=12)
plt.xticks(k_values)
plt.ylim(50, 105)  # 设定合理的Y轴范围
plt.grid(True, linestyle='--', alpha=0.6)

# 添加图例
plt.legend()

# 显示图像
plt.show()
```
